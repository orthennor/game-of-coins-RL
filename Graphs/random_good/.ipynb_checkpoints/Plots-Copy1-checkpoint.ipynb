```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from scipy.interpolate import spline
from tqdm import tqdm
```

```python
data_1 = pd.read_csv("rewards_as_their_bad.csv")
```

```python
data_1.columns
```

```python
reward_1 = data_1['50 wo map - reward']
numbers_1 = data_1['episode']
```

```python
len(numbers_1)
```

```python
def smooth(values, w_range):
    new_values = []

    for val in tqdm(range(len(values))):
        
        w_start = val - w_range
        w_end = val + w_range
        w_start = max(w_start, 0)
        w_end = min(w_end, (len(values) - 1))
        loc_vals = values[w_start : w_end]
        local_value = sum(loc_vals)/len(loc_vals)
        new_values.append(local_value)
    return new_values
```

```python
smoothed_reward_1 = smooth(reward_1, 100)
```

```python
plt.figure(figsize=(20, 10), dpi = 300)
plt.plot(numbers_1, list(reward_1), alpha=0.2, linewidth=0.5)
plt.plot(numbers_1, smoothed_reward_1, label='reward', linewidth=1.25)
plt.xlabel('episode', fontsize=40)
plt.ylabel('reward', fontsize=40)
plt.xticks(fontsize=25, rotation=0)
plt.yticks(fontsize=25, rotation=0)
plt.grid(linewidth='0.5')
plt.legend(fontsize=40)
plt.show()
```

```python
c1_n = pd.read_csv("neither_coins_as_their_bad.csv")
c1_o = pd.read_csv("other_coins_as_their_bad.csv")
c1_s = pd.read_csv("self_coins_as_their_bad.csv")
```

```python
episodes_1 = c1_n['episode']
c1_neth = c1_n['50 wo map - neither_coins2']
c1_othe = c1_o['50 wo map - other_coins2']
c1_self = c1_s['50 wo map - self_coins2']
```

```python
smoothing = 1000
c1_neth_smoth = smooth(c1_neth, smoothing)
c1_othe_smoth = smooth(c1_othe, smoothing)
c1_self_smoth = smooth(c1_self, smoothing)
```

```python
plt.figure(figsize=(20, 10), dpi = 300)
#plt.plot(episodes_1, list(c1_neth), alpha=0.2, c='g', linewidth=0.5)
plt.plot(episodes_1, c1_neth_smoth, label='neither coins', c='g', linewidth=1.2)

#plt.plot(episodes_1, list(c1_othe), alpha=0.2, c='r', linewidth=0.5)
plt.plot(episodes_1, c1_othe_smoth, label='other coins', c='r', linewidth=1.2)

#plt.plot(episodes_1, list(c1_self), alpha=0.2, c='b', linewidth=0.5)
plt.plot(episodes_1, c1_self_smoth, label='self coins', c='b', linewidth=1.2)
plt.xlabel('episode', fontsize=40)
plt.ylabel('coins', fontsize=40)
plt.xticks(fontsize=25, rotation=0)
plt.yticks(fontsize=25, rotation=0)
plt.grid(linewidth='0.5')
plt.legend(fontsize=30)
plt.show()
```
