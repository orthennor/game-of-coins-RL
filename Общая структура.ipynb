У нас есть одна общая сетка f которая состоит из LSTM и двух полносвязанных слоев с ELU активациями. Также для каждого агента у нас есть свои softmax слой который принимает выход из сетки f и возвращает распределение по действиям и линейный слой который также принимает выход из f и возвращает значение для value функции. Все эти слои мы во время игры не обновляем. f на вход получает state из среды для агента, его goal и вектор z_other для другого агента. Вот этот вектор z_other для  мы и будем обновлять по ходу одного эпизода. Мы сначала берем для каждого агента совершаем действие. Потом каждый агент из своих данных(s, z, z_other) пытается прикинуть, какое действие совершил бы другой агент (s_other, argmax(z_other), z), получаем распределение действий, берется истинное действие для данного агента, считается cross_entropy_loss, считаются градиенты для z_other каждого агента, делаем шаг оптимизации. Параметры сетки f, policy и value обновляются вне игры с A3C.
